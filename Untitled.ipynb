```python
import pandas as pd
import matplotlib.pyplot as plt
import mpld3
df = pd.read_csv('water_data.csv')
pd.set_option('display.max_columns', None)
df = df.drop(['Reference Description','Indicator Value','Description','Update Frequency','Method','Y axis','Indicator Name','Reference time','Reference value','Rule','Sub-AOI','Color code'],axis =1)
df[["date","Time"]] = df["Time"].str.split("T", expand = True)
df[["Hour", "Minute", "Sec"]] = df["Time"].str.split(":", expand = True)
df[["Year", "Month", "Day"]] = df["date"].str.split("-", expand = True)
df[["Latitude", "Longitude"]] = df["AOI"].str.split(",", expand = True)
df = df.drop(['date','Time','AOI'],axis = 1)
df = df.dropna()
df.rename(columns = {'Site Name': 'Site_Name', 'EO Sensor': 'EO_Sensor', 'Input Data': 'Input_Data', 'Indicator code': 'Indicator_code', 'Measurement Value': 'Measure', 'Data Provider': 'Data_Provider'}, inplace = True)

d = df.loc[df['AOI_ID'] == 'US04']
fig = plt.figure(figsize = (18,8))
plt.subplot(2,2,1)


plt.bar(d['Year'],d['Measure'])

plt.title('Water Quality Value in the Pandemic IN New York ')
plt.xlabel('Years')
plt.ylabel('Water Quality')


plt.subplot(2,2,2)


plt.bar(df['Country'],df['Measure'])


plt.title('Water Quality Value in the Pandemic ')
plt.xlabel('Data input')
plt.ylabel('Water Quality')

plt.subplot(2,2,3)




plt.bar(df['EO_Sensor'],df['Measure'])


plt.title('Water Quality Value Dependency on Sensor device ')
plt.xlabel('EO Sensor')
plt.ylabel('Water Quality')


plt.subplot(2,2,4)


plt.bar(df['Region'],df['Measure'])


plt.title("Region wise VS Measurement Value")
plt.ylabel("Water Quality values")
plt.xlabel("Regions")

plt.tight_layout()

```

```python
html_str = mpld3.fig_to_html(fig)
Html_file= open("final_data.html","w")
Html_file.write(html_str)
Html_file.close()
```
